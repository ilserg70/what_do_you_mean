## MatPlot examples

```python
import math
t = list(range(360))
x = [i * 2* math.pi/360 for i in t]
y1 = [math.sin(a) for a in x]
y2 = [math.cos(a) for a in x]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x, y1, 'r--', x, y2, 'g--')
plt.title('sin(x) & cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
